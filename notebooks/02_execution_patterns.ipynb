# Execution patterns

Patterns describe how shared components exchange control. The cells call the repository's six existing deterministic implementations rather than reproducing their orchestration logic.

```python
from agentic_tutorial.education import PATTERN_NAMES, run_pattern_async

pattern_results = {name: await run_pattern_async(name) for name in PATTERN_NAMES}
[(name, result["pattern"], result["limitation"]) for name, result in pattern_results.items()]
```

Prompt chaining passes a bounded artefact through fixed stages. Routing selects work, while parallelisation runs independent workers and restores deterministic ordering. ReAct interleaves reasoning, action and observation.

```python
assert len(pattern_results["prompt-chaining"]["stages"]) == 3
assert pattern_results["routing-parallelisation"]["ordered_results"] == [
    "evidence",
    "claims:complete",
    "metadata:complete",
]
assert pattern_results["react-tool-use"]["success"] is True
```

Planner–executor separates task decomposition from bounded execution. Critic–reviser permits a finite correction cycle. Orchestrator–worker delegates to specialists with genuinely different tool permissions.

```python
assert pattern_results["planner-executor"]["completed"] == ["search", "extract", "synthesise"]
assert pattern_results["critic-reviser"]["revisions"] == 1
assert pattern_results["orchestrator-worker"]["separate_permissions"] is True
pattern_results
```

No pattern is universally best: each result exposes a limitation, and every loop remains bounded by construction.
